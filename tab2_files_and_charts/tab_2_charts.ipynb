```r
library(ggplot2)
```

```r
chart_11 <- read.csv('chart_11.csv')
chart_12 <- read.csv('chart_12.csv')
chart_13 <- read.csv('chart_13.csv')

chart_12$Season <- as.factor(chart_12$Season)
chart_13$Season <- as.factor(chart_13$Season)
```

## Chart 11

```r
ggplot(subset(chart_11, Player == 'Kobe Bryant' & Stage == 'Regular_Season')) +
    geom_line(aes(x = Season, y = per_game, color = TWO_PT_3PT_eFG), size = 2)  + 
    scale_x_continuous(labels = as.character(chart_11$Season), breaks = chart_11$Season) +
    ylab('Shooting Percentage') +
    ggtitle('Average Shooting Percentages by Season') +
    scale_color_manual("TWO_PT_3PT_eFG", values = c('2PT_%' = 'steelblue3', '3PT_%' = 'darkorange', 'eFG_%' = 'coral2')) + 
    theme(
    axis.text.x = element_text(angle = 90, size = 13),
    axis.text.y = element_text(size = 13),
    legend.position = "bottom",
    legend.justification = "left",
    legend.text = element_text(size = 10),
    legend.title = element_blank(),
    plot.title = element_text(hjust = 0.5, size = 18),
    panel.grid.major.x=element_blank(),
    panel.grid.minor.x=element_blank(),
    panel.grid.major.y=element_line(colour = 'lightgrey'),
    panel.grid.minor.y=element_blank(),
    panel.background=element_rect(fill="white"),
    axis.line = element_line(colour = 'black'),
    axis.title.x = element_text(vjust=-0.5, size = 15), axis.title.y=element_text(vjust=0.1, size = 15)
    )
```

## Chart 12

```r
ggplot(subset(chart_12, Player == 'Kobe Bryant' & Stage == 'Regular_Season')) +
    geom_col(aes(x = Season, y = True.shooting.percentage), fill = 'steelblue4') + 
    ylab('True Shooting Percentage') +
    ggtitle('True Shooting Percentage by Season') +
    theme(
    axis.text.x = element_text(angle = 90, size = 13),
    axis.text.y = element_text(size = 13),
    plot.title = element_text(hjust = 0.5, size = 18),
    panel.grid.major.x=element_blank(),
    panel.grid.minor.x=element_blank(),
    panel.grid.major.y=element_line(colour = 'lightgrey'),
    panel.grid.minor.y=element_blank(),
    panel.background=element_rect(fill="white"),
    axis.line = element_line(colour = 'black'),
    axis.title.x = element_text(vjust=-0.5, size = 15), axis.title.y=element_text(vjust=0.1, size = 15)
    )
```
